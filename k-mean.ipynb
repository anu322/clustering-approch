```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('..//DataSets/Mall_Customers.csv')
df
```

```python
x = df.iloc[:, [3, 4]].values
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('kmean clustering')
plt.xlabel('numbers of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
y_pred_mean=kmeans.fit_predict(x)
y_pred_mean
```

```python
plt.scatter(x[y_pred_mean==0,0],x[y_pred_mean==0,1],s=100,c='b',label='miser')
plt.scatter(x[y_pred_mean==1,0],x[y_pred_mean==1,1],s=100,c='y',label='aware')
plt.scatter(x[y_pred_mean==2,0],x[y_pred_mean==2,1],s=100,c='r',label='target')
plt.scatter(x[y_pred_mean==3,0],x[y_pred_mean==3,1],s=100,c='k',label='careless')
plt.scatter(x[y_pred_mean==4,0],x[y_pred_mean==4,1],s=100,c='g',label='careful')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200, c='magenta',label='centeroids')
plt.title('k-mean-clustering')
plt.xlabel('anaual income')
plt.ylabel('speding score')
plt.legend()
plt.show()

```
